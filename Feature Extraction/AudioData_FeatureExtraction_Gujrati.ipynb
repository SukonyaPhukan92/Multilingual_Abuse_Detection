# **2. Audio Feature Extraction using wavLM**

```python
import torchaudio
```

```python
!pip install transformers
```

```python
from transformers import AutoProcessor, WavLMModel, Wav2Vec2Processor
processor = Wav2Vec2Processor.from_pretrained("patrickvonplaten/wavlm-libri-clean-100h-base-plus")
wavlm = WavLMModel.from_pretrained("patrickvonplaten/wavlm-libri-clean-100h-base-plus")# audio file is decoded on the fly

def extract_features(path):
    sample_rate = 16000
    array, fs = torchaudio.load(path)
    input = processor(array.squeeze(), sampling_rate= sample_rate, return_tensors="pt")
    # apply the model to the input array from wav
    with torch.no_grad():
       outputs = wavlm(**input)
    # extract last hidden state, compute average, convert to numpy
    last_hidden_states = outputs.last_hidden_state.squeeze().mean(axis=0).numpy()
    return last_hidden_states
```

## **2.2.1 Dataset loading**

```python
csv_file_path = '/content/drive/MyDrive/Audio_Data/annotations/annotations/Gujarati_train.csv'
# Read the CSV file into a DataFrame
guj_train = pd.read_csv(csv_file_path)
guj_train.head()
```

## **2.2.2 Feature Extraction**

```python
image_directory_guj = '/content/drive/MyDrive/Audio_Data/Indian_Language_Audio_Dataset/SC_audio_Gujarati/'
guj_train['filename']= image_directory_guj + "/" + guj_train['filename']
guj_train.head()
```

```python
#save embedding in Numpy
```

```python
guj_train_embeddings = []
for i in range(len(guj_train)):
    features = extract_features(guj_train['filename'][i])
    guj_train_embeddings.append(features)

guj_train_embeddings = np.array(guj_train_embeddings)
print(guj_train_embeddings.shape)
```

```python
np.save('/content/drive/MyDrive/Audio_Data/Output/guj_train_embeddings.npy',guj_train_embeddings)

print("guj_train_embeddings.npy file is saved")
```
